- Sqft_above: square footage of house apart from the basement
- Sqft_living15: Living room area in 2015(implies-- some renovations) This might or might not have affected the lotSize area.
- Sqft_lot15: lotSize area in 2015(implies-- some renovations)

-------------------------------------------------------------------------------------------------
The following columns are used as a classification feature, so it would make sense to convert them into category:
- waterfront
- condition
- grade

And since id is the unique key, we shouldn't see it necessarily as a numerical column, therefore we can also convert it into type category.

------------------------------------------------------------------------------------------------

```python
# List of columns to convert to categorical
columns_to_convert = ['waterfront', 'condition', 'grade', 'id']

# Convert specified columns to categorical
for col in columns_to_convert:
    df[col] = df[col].astype('category')
    
# Display the data types of the DataFrame
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt

# Get the names of the categorical columns
categorical_columns = [col for col in df.columns if pd.api.types.is_categorical_dtype(df[col]) and col != 'id']

# Generate bar plots for each categorical column (excluding 'Id')
for col in categorical_columns:
    df[col].value_counts().plot(kind='bar')
    plt.title('Bar plot for ' + col)
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
import seaborn as sns

for col in categorical_columns:
  sns.barplot(x=df[col], y=df["price"])
  plt.xticks(rotation=90)
  plt.show()
```
